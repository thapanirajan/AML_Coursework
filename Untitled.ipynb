## Text preprocessing 

```python
# Load stop words 
stop_words = set(stopwords.words('english'))
```

```python
# Initialize Lemmatzer converts words like (running, runs, ran) to run 
lemmatizer = WordNetLemmatizer()
```

```python
# helper function to preprocess text 
def preprocess_text(text):
    """
    Clean and preprocess input text for NLP tasks.
    """
    # Convert text to lowercase
    text = text.lower()
    
    # Remove punctuation, special characters, and numbers
    text = re.sub(r'[^a-z\s]', '', text)
    
    # Tokenize text
    tokens = text.split()
    
    # Remove stopwords and apply lemmatization
    cleaned_tokens = [
        lemmatizer.lemmatize(word)
        for word in tokens
        if word not in stop_words
    ]
    
    return " ".join(cleaned_tokens)
```

```python
# applying helper function to dataset
train_df['clean_text'] = train_df['Text'].apply(preprocess_text)
```

```python
train_df[['Text', 'clean_text']].head()
```

```python
train_df['clean_word_count'] = train_df['clean_text'].apply(lambda x: len(x.split()))
train_df[['word_count', 'clean_word_count']].describe()
```

## Feature Extraction 

```python

from sklearn.model_selection import train_test_split

X = train_df['clean_text']
y = train_df['Category']

X_train, X_val, y_train, y_val = train_test_split(X,y,test_size=0.2)

print("Training samples:", X_train.shape[0])
print("Validation samples:", X_val.shape[0])
```

```python
tfidf = TfidfVectorizer(max_features=5000, ngram_range=(1,2))
```

```python
# Fit TF-IDF on training data and transform both train and test data
X_train_tfidf = tfidf.fit_transform(X_train)
X_val_tfidf = tfidf.transform(X_val)
```

```python
# shape of X_train and X_test 
print("Training shape:", X_train_tfidf.shape)
print("Testing shape:", X_val_tfidf.shape)
```

## Model Training 

## Model 1 – Multinomial Naïve Bayes

```python
# initialize model 
nb_model = MultinomialNB()
```

```python
# model train 
nb_model.fit(X_train_tfidf, y_train)
```

```python
# Prediction 
nb_predictions = nb_model.predict(X_val_tfidf)
```

```python
nb_predictions
```

## Performance Evaluation for Multinomial Naïve Bayes

```python
# accuracy 
nb_accuracy = accuracy_score(y_val, nb_predictions)
nb_accuracy
```

## Classification Report 

```python
print(classification_report(y_val, nb_predictions))
```

```python
cm_nb = confusion_matrix(y_val, nb_predictions)
```

```python
print(train_df['Category'].value_counts())
print("Confusion matrix:")
print(cm_nb)
```

```python
plt.figure()
sns.heatmap(
    cm_nb,
    annot=True,
    fmt='d',
    cmap='Blues',
    xticklabels=nb_model.classes_,
    yticklabels=nb_model.classes_
)
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix - Multinomial Naïve Bayes')
plt.show()
```

## Model Training 2 – Logistic Regression

```python
# n_jobs is set to -1 for parallel processing , max_iter is set to 1000 so that model can run upto 1000 optimization steps to find best parameter.
lr_model = LogisticRegression(max_iter=1000, n_jobs=-1)
```

```python
# model training 
lr_model.fit(X_train_tfidf, y_train)
```

```python
# make prediction 
lr_prediction = lr_model.predict(X_val_tfidf)
```

```python
lr_prediction
```

## Performance evaluation 

```python
lr_accuracy = accuracy_score(y_val, lr_prediction)
lr_accuracy
```

## CLassification Report 

```python
print(classification_report(y_val, lr_prediction))
```

## Confusion matrix 

```python
print(train_df['Category'].unique())
cm_lr = confusion_matrix(y_val, lr_prediction)
print("Confusion matrix")
print(cm_lr)
```

## Model Comparison

```python
comparison_df = pd.DataFrame({
    "Model": ["Multinomial Naïve Bayes", "Logistic Regression"],
    "Accuracy": [nb_accuracy, lr_accuracy]
})

comparison_df
```

```python
# Extracting evaluation matrix 
nb_report = classification_report(y_val, nb_predictions, output_dict=True)
lr_report = classification_report(y_val, lr_prediction, output_dict=True)
```

```python
comparison_metrics = pd.DataFrame({
    "Model": ["Multinomial Naïve Bayes", "Logistic Regression"],
    "Accuracy": [nb_accuracy, lr_accuracy],
    "Macro Precision": [
        nb_report["macro avg"]["precision"],
        lr_report["macro avg"]["precision"]
    ],
    "Macro Recall": [
        nb_report["macro avg"]["recall"],
        lr_report["macro avg"]["recall"]
    ],
    "Macro F1": [
        nb_report["macro avg"]["f1-score"],
        lr_report["macro avg"]["f1-score"]
    ],
    "Weighted F1": [
        nb_report["weighted avg"]["f1-score"],
        lr_report["weighted avg"]["f1-score"]
    ]
})

comparison_metrics
```

```python
classes = list(nb_report.keys())[:-3] 

class_f1_df = pd.DataFrame({
    "Category": classes,
    "Naïve Bayes F1": [nb_report[c]["f1-score"] for c in classes],
    "Logistic Regression F1": [lr_report[c]["f1-score"] for c in classes]
})

class_f1_df
```

```python
plt.figure(figsize=(12,10))
class_f1_df.set_index("Category").plot(kind="bar")
plt.title("Class-wise F1 Score Comparison")
plt.ylabel("F1 Score")
plt.ylim(0.9, 1.0)
plt.xticks(rotation=45)
plt.show()
```
